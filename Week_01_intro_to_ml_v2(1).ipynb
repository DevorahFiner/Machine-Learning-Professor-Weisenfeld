# A Brief History of ML
- **1943** Logician Walter Pitts and neuroscientist Warren McCulloch published the first mathematical modeling of a neural network to create algorithms that mimic human thought processes.
- **1951** Marvin Minsky and Dean Edmonds developed the first artificial neural network (ANN) called SNARC using 3,000 vacuum tubes to simulate a network of 40 neurons.
- **1958** Frank Rosenblatt developed the perceptron, an early ANN that could learn from data and became the foundation for modern neural networks.
- **1959** Arthur Samuel coined the term machine learning in a seminal paper explaining that the computer could be programmed to outplay its programmer.
- **1967** The nearest neighbor algorithm provided computers with the capability for basic pattern recognition and was used by traveling salespeople to plan the most efficient routes via the nearest cities.
- **1969** Arthur Bryson and Yu-Chi Ho described a backpropagation learning algorithm to enable multilayer ANNs, an advancement over the perceptron and a foundation for deep learning.


- **1989** Yann LeCun, Yoshua Bengio and Patrick Haffner demonstrated how convolutional neural networks (CNNs) can be used to recognize handwritten characters, showing that neural networks could be applied to real-world problems.
- **1997** Sepp Hochreiter and Jrgen Schmidhuber proposed the Long Short-Term Memory recurrent neural network, which could process entire sequences of data like speech or video.
- **1997** IBM's Deep Blue defeated Garry Kasparov in a historic chess rematch, the first defeat of a reigning world chess champion by a computer under tournament conditions.
- **2006** IBM Watson originated with the initial goal of beating a human on the Jeopardy! quiz show. In 2011, the question-answering computer system defeated the show's all-time (human) champion Ken Jennings.
- **2017** Google researchers developed the concept of transformers in the seminal paper "Attention is all you need," inspiring subsequent research into tools that could automatically parse unlabeled text into large language models (LLMs).
- **2018** OpenAI released GPT (Generative Pre-trained Transformer), paving the way for subsequent LLMs.
- **2022** OpenAI released ChatGPT in November to provide a chat-based interface to its GPT 3.5 LLM.
- **2023** OpenAI announced the GPT-4 multimodal LLM that receives both text and image prompts.

## Supervised vs. Unsupervised Learning

### Supervised Learning
- Supervised learning is trying to learn how to predict labels (the answer) from features (clues) by learning patterns data with known labels
- The goal is that the patterns we learn from the known data will generalize to data with unknown labels
- Most of what we study will be supervised learning, but we'll do some unsupervised
- Example: Studying for a standardized test by doing practice exams and checking your answers afterward.


### Unsupervised Learning
- In Unsupervised learning, there are no "correct" labels. We are trying to learn something about the data itself that is implicitly but not explicitly encoded in the data.
- The goal is to create structure around the data that can potentially be valuable on it's own, or sometimes as features for a supervised model
- Example: Organizing your closet by color.

## Mistake Bound Learning - The Perceptron
One of the earliest (and simplest) binary classification algorithms - only works on linearly separable data (or data that can be made linearly separable) (Linearly Separable = classes that can be fully separated by a hyperplane)

Data: 
- $X = $ features ($n \times d$ matrix) ($x_i$ is a $d$-length feature vector representing the $i$th row of $X$), 
- $y \in \{0,1\} = $ target ($n$-length target vector of binary values corresponding to each row of $X$) ($y_i$ is the $i$th target value)

Learnable Parameters:
- $w = $ weights ($d$-length weight vector to be learned by algorithm) ($w_j$ is the $j$th weight value)
- note, you can explicitly add a bias parameter, but we can simply add that implicitly by adding a column of just $1$'s to $X$ and increasing the length of $w$ by $1$

**Perceptron**: $\hat{y} = \Theta(Xw)$

Where: 
- $\hat{y} = $ the predicted values of $y$
- $\theta(z)=\begin{cases}
      1, & \text{if}\ z>0 \\
      0, & \text{if}\ z\leq0
    \end{cases}$ - AKA: Heaviside Step Function (note that above we used $\Theta$ - the vector version, and here we are using $\theta$ - the scalar version)

Reminder: 
- $Xw$ is the product of $n \times d$ matrix $X$ and $d$-length vector $w$ ($w$ can also be thought of as a $d \times 1$ matrix)
- In matrix multiplication such as $C = AB$, the number of columns in $A$ must match the number of rows in $B$, and $C$ will have the same number of rows as $A$ and the same number of columns as $B$. So if $A$ is $m\times n$ and $B$ is $n\times p$, then $C$ will be $m \times p$ 

## The Perceptron Training Algorithm:
1. Initilize $w$ as a $d$-length vector of $0$'s
2. For each $x_i$ compute: $\hat{y_i} = \theta(x_i^\top w)$
3. if  $\hat{y_i} \neq y_i$: update $w = \begin{cases}
                                                  w + x_i, & \text{if}\ y_i = 1 \\
                                                  w - x_i, & \text{if}\ y_i = 0\\
                                               \end{cases}$ (do not change $w$ if $y_i = y$)
4. repeat from step 2 until no more mistakes made

If the data is linearly separable, there exists a theoretical hyperplane that maximizes the margin $\gamma$, the minimum distance to any point. If we encapsulate all data points within a hypersphere of radius $R$, it is guaranteed that the perceptron algorithm will make at most $\frac{R}{\gamma^2}$ mistakes.

(Image Source: https://www.cs.cmu.edu/~mgormley/courses/606-607-f18/slides607/lecture4-pmb.pdf)

Example: Let's look at this penguin data:

```python
df = sns.load_dataset('penguins')
df = df[df['species'].isin(['Adelie', 'Gentoo'])].dropna().reset_index()
df = df[['bill_depth_mm', 'flipper_length_mm', 'species']]
pd.concat([df.head(2), df.tail(2)])  
```

```python
X, y = df[['bill_depth_mm', 'flipper_length_mm']], df['species']
sns.scatterplot(X, x='bill_depth_mm', y='flipper_length_mm', hue=y);
```

As you can see, the two species are linearly separable with these two features.

Let's write the Perceptron Code

```python
class Perceptron:    
    
    def fit(self, X, y, max_epochs = 1000):
        X, y = np.array(X), np.array(y) # convert to numpy arrays
        self.w = np.zeros(X.shape[1]) # initialize w as all 0's
        self.mistakes, self.epochs = 0, 0 # initialize mistake & epoch count (for informational purposes)
        for epoch in range(max_epochs): #let's not go on forever if not linearly separable
            self.epochs += 1 # increment the epoch count
            epoch_errors = 0 # initialize the count or errors in this epoch
            for i in range(X.shape[0]): # for each row of X
                error = y[i] - int(X[i] @ self.w > 0)  # error can be -1, 0, or 1 
                if error != 0: 
                    epoch_errors += 1
                    self.w += error * X[i] 
            if epoch_errors == 0: break # we made no errors, so we can end this function
            self.mistakes += epoch_errors # update the global errors
        return self
            
    def predict(self, X) -> np.array:
        return (np.array(X) @ self.w > 0).astype(int)        
```

Let's test it out!

```python
y = df['species'].map({'Gentoo': 1, 'Adelie': 0}) # map to 1's and 0's
perceptron = Perceptron().fit(X, y)

print(f"weights={perceptron.w}, mistakes={perceptron.mistakes}, epochs={perceptron.epochs}")

y_hat = pd.Series(perceptron.predict(X)).map({1: 'Gentoo', 0: 'Adelie'})
sns.scatterplot(X, x='bill_depth_mm', y='flipper_length_mm', hue=y_hat);
m = -perceptron.w[0] / perceptron.w[1] # slope of dividing line
plt.plot([14, 20], [ m * 14, m * 20], color = 'black');
```

The black line is a separating hyperplane found by taking the orthogonal vector to $w$.

## The k Nearest Neighbors Model
A simple model to estimate the value of an unknown data point by looking the values of its nearest $k$ data points and taking their average.

- The number $k$ is selected in advance (you can test multiple $k$'s)
- Distance can be computed in multiple ways, often it can be the Euclidean distance, but Manhattan distance is also an option
- Various averages can be used, typically it is the mean, but it could be median, mode, or even a weighted averge
- Very sensitive to feature scale and training set size.
- The ML implementation of the principle "Birds of a feather flock together"

Let's look at an Example.

Below are a set of geographic points near Denver, Colorado. Our data set consists of each point's coordinates and elevation in meters above sea level.

```python
df_train = pd.read_csv('elev_train.csv')
df_train.head()
```

The colors of the dots represent that point's elevation in meters above sea level.

```python
df_train = pd.read_csv('elev_train.csv')

ax = sns.scatterplot(df_train, x='lon', y='lat', hue='elevation', palette = sns.color_palette("mako", as_cmap=True))
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

We have a set of orange points, for which we'll pretend we don't know the elevation.

```python
df_valid = pd.read_csv('elev_valid.csv')
ax = sns.scatterplot(df_train, x='lon', y='lat', hue='elevation', palette = sns.color_palette("mako", as_cmap=True))
ax = sns.scatterplot(df_valid, x='lon', y='lat', c='orange')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

Is it reasonable to assume that the orange dots probably have elevations similar to the nearby bluish dots?

Let's train a KNN model to guess the elevation of the orange points based on the elevation of the nearest known points.

```python
from sklearn.neighbors import KNeighborsRegressor

X_train, y_train = df_train[['lat', 'lon']], df_train['elevation']

knn = KNeighborsRegressor(5).fit(X_train, y_train)

X_valid, y_valid = df_valid[['lat', 'lon']], df_valid['elevation']

print(f"R-squared: {knn.score(X_valid, y_valid)}")

y_hat = knn.predict(X_valid)

sns.scatterplot(pd.DataFrame({'y_true': y_valid, 'y_pred': y_hat}), x='y_true', y='y_pred')
plt.plot([min(y_valid), max(y_valid)], [min(y_valid), max(y_valid)], color = 'black');
```

The black line represents a perfect prediction

The algorithm is very sensitive to k.

```python
ks = list(range(1, 21))
r2s = [KNeighborsRegressor(k).fit(X_train, y_train).score(X_valid, y_valid) for k in ks]

sns.set_style("darkgrid")
ax = sns.lineplot(pd.DataFrame({'k': ks, 'R2': r2s}), x='k', y='R2')
ax.set_xticks(ks);
```
