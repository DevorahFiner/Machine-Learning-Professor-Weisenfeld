```python
import gradio as gr
from diffusers import DiffusionPipeline

# Load the pre-trained Stable Diffusion model from Hugging Face
pipeline = DiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")

def generate_image(prompt):
    """
    Generate an image based on the given text prompt using the pre-loaded pipeline.

    Args:
    prompt (str): The text prompt to generate an image from.

    Returns:
    PIL.Image.Image: The generated image.
    """
    # Generate an image based on the text prompt
    image = pipeline(prompt).images[0]
    return image

# Create a Gradio interface for the image generation function
demo = gr.Interface(
    fn=generate_image,            # The function to call for image generation
    inputs="text",                # Input component type (text)
    outputs="image",              # Output component type (image)
    title="Text to Image Generator",  # Title of the Gradio interface
    description="Enter a text prompt to generate an image using the Stable Diffusion model."  # Description of the interface
)

# Launch the Gradio interface and create a public shareable link
demo.launch(share=True)
```
